```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
os.chdir('D:\dataAnalysisProjects\project5')
```

```python
salesOrders = pd.read_excel('Sales Analysis Report.xlsx', sheet_name=0)
dimCustomers = pd.read_excel('Sales Analysis Report.xlsx', sheet_name=1)
dimRegions = pd.read_excel('Sales Analysis Report.xlsx', sheet_name=2)
dimProducts = pd.read_excel('Sales Analysis Report.xlsx', sheet_name=3)
salesOrders.head()
```

```python
mapping_names = {k:v for k, v in zip(dimCustomers['Customer Index'], dimCustomers['Customer Names'])}
mapping_products = {k:v for k, v in zip(dimProducts['Index'], dimProducts['Product Name'])}
salesOrders['Customer Name Index'] = salesOrders['Customer Name Index'].map(mapping_names)
salesOrders['Product Description Index'] = salesOrders['Product Description Index'].map(mapping_products)
salesOrders.head()
```

```python
salesOrders['City'] = salesOrders['Delivery Region Index']
mapping_cities = {k:v for k, v in zip(dimRegions['Index'], dimRegions['City'])}
salesOrders['City'] = salesOrders['City'].map(mapping_cities)
salesOrders.head()
```

```python
mapping_suburbs = {k:v for k, v in zip(dimRegions['Index'], dimRegions['Suburb'])}
salesOrders['Delivery Region Index'] = salesOrders['Delivery Region Index'].map(mapping_suburbs)
salesOrders.head()
```

```python
salesOrders['Currency Code'].unique()
```

```python
def unifiedCurrency(currency, price):
    if currency == 'NZD':
        return price*.6
    elif currency == 'AUD':
        return price*.64
    elif currency == 'GBP':
        return price*1.23
    elif currency == 'EUR':
        return price*1.06
    else:
        return price
```

```python
salesOrders['Unit Selling Price Unified'] = salesOrders.apply(lambda row: unifiedCurrency(row['Currency Code'], row['Unit Selling Price']), axis=1)
salesOrders['Unit Cost Unified'] = salesOrders.apply(lambda row: unifiedCurrency(row['Currency Code'], row['Unit Cost']), axis=1)
salesOrders.head()
```

```python
salesOrders.drop(['Currency Code', 'Unit Selling Price', 'Unit Cost'], inplace=True, axis=1)
salesOrders.head()
```

```python
salesOrders.columns = ['order_number', 'order_date', 'ship_date', 'customer_name', 'channel', 'warehouse_code', 'suburb', 'product', 'order_quantity', 'city', 'unit_price', 'unit_cost']
salesOrders.head()
```

```python
salesOrders['revenue'] = (salesOrders['unit_price']-salesOrders['unit_cost']) * salesOrders['order_quantity']
salesOrders.head()
```

```python
for col in salesOrders.select_dtypes(include='float'):
    sns.histplot(data=salesOrders, x=col, color='grey', alpha=.2)
    plt.axvline(x=salesOrders[col].median(), color='black', linewidth = 1.8)
    plt.axvline(x=salesOrders[col].mean(), color='red', linewidth = 1.8)
    plt.axvline(x=salesOrders[col].quantile(.95), color='blue', linewidth=1.8)
    sns.despine()
    plt.show()
# all of the 3 histos are left skewed => more likely to have low values => median < mean
```

```python
# month rolling average:
sales_orders_copy = salesOrders.copy(deep=True)
sales_orders_copy = sales_orders_copy.set_index('order_date')
sales_orders_copy_resampled = sales_orders_copy[['revenue']].resample('M').sum()
ax = sales_orders_copy_resampled.plot()
plt.text(x='2017-8', y=1000000, s='South Sandwich Trench', fontsize=12)
plt.text(x='2019-7', y=600000, s='Great, It\'s Mariana Trench now', fontsize=12)
sns.despine()
plt.show()

# The sales decrease rapidly before and after December, I think it's because people tend to buy at christmas because of generous offers.
# but still something suspicious, it decreases also at the middle of the year, look at Jun 2017 and Jul 2019!
```
